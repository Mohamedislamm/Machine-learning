# Task
Scrape text content from "https://en.wikipedia.org/wiki/Data_science" and "https://en.wikipedia.org/wiki/Machine_learning", then perform text preprocessing by converting to lowercase, removing special characters, numbers, and punctuation, tokenizing, and removing NLTK stopwords. Finally, identify all unique words across both datasets and summarize the results by reporting the total number of unique terms found.

```python
import requests, re, nltk
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Setup and requirements
nltk.download(['punkt', 'punkt_tab', 'stopwords'], quiet=True)
website_link_1 = "https://en.wikipedia.org/wiki/Data_science"
website_link_2 = "https://en.wikipedia.org/wiki/Machine_learning"

def get_text(url):
    res = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
    soup = BeautifulSoup(res.text, 'html.parser')
    return ' '.join([p.text for p in soup.find_all('p')])

# 1. Scrape Data
text = get_text(website_link_1) + ' ' + get_text(website_link_2)

# 2. Lowercasing
text = text.lower()

# 3. Removing Special Characters, Numbers and Punctuation
text = re.sub(r'[^a-z\s]', '', text)

# 4. Tokenization
tokens = word_tokenize(text)

# 5. Stopword Removal
stop_words = set(stopwords.words('english'))
filtered = [w for w in tokens if w not in stop_words]

# 6. Get Unique Values
unique_words = sorted(set(filtered))

print(f"Total unique words found: {len(unique_words)}")
print(f"First 50 unique words: {unique_words[:50]}")
```
